```python
%%writefile Workout.h
#ifndef WORKOUT_H // include guard
#define WORKOUT_H

#include <string> // for string class
using namespace std;

class Workout {
private:                        // private data members and attributes of the Workout class
  string activity_;
  int intensityLevel_;
  int caloriesBurnPerMinute_;

public:
  // constructors
  Workout();
  Workout(const string& activity, int intensityLevel, int caloriesBurnPerMinute);

  // getters
  string getActivity() const;
  int getIntensityLevel() const;
  int getCaloriesBurnPerMinute() const;

};

#endif // WORKOUT_H
```

```python
%%writefile Workout.cpp

#include "Workout.h"

// Default Constructor
Workout::Workout() {
  activity_ = "";
  intensityLevel_ = 0;
  caloriesBurnPerMinute_ = 0;
}

// Overloaded Constructor
Workout::Workout(const string& activity, int intensityLevel, int caloriesBurnPerMinute) {
  activity_ = activity;
  intensityLevel_ = intensityLevel;
  caloriesBurnPerMinute_ = caloriesBurnPerMinute;
}

// Getters
string Workout::getActivity() const {
  return activity_;
}

int Workout::getIntensityLevel() const {
  return intensityLevel_;
}

int Workout::getCaloriesBurnPerMinute() const {
  return caloriesBurnPerMinute_;
}
```

```python
%%writefile BodyMetric.h
#ifndef BODYMETRIC_H // include guard
#define BODYMETRIC_H

#include <string> // for string class
using namespace std;

class BodyMetric {
private:
  int age_;
  string gender_;
  float weight_;
  float height_;
  string goal_;
  float bmi_;
  string bmiStatus_;

public:

  // constructors
  BodyMetric();
  BodyMetric(int age, string gender, float weight, float height, string goal);

  // BMI operations
  float calculateBMI() const;
  void updateBMI();
  string determineBMIStatus();

  // Getters
  int getAge() const;
  string getGender() const;
  float getWeight() const;
  float getHeight() const;
  string getGoal() const;
  float getBMI() const;
  string getBMIStatus() const;

  // Setters
  void setAge(int age);
  void setGender(string gender);
  void setWeight(float weight);
  void setHeight(float height);
  void setGoal(string goal);

};

#endif // BODYMETRIC_H
```

```python
%%writefile BodyMetric.cpp

#include <string>
#include "BodyMetric.h"
using namespace std;

// Default Constructor
BodyMetric::BodyMetric() {
  age_ = 0;
  gender_ = "";
  weight_ = 0.0;
  height_ = 0.0;
  goal_ = "";
}

// Overloaded Constructor
BodyMetric::BodyMetric(int age, string gender, float weight, float height, string goal) {
  age_ = age;
  gender_ = gender;
  weight_ = weight;
  height_ = height;
  goal_ = goal;
}

// BMI opeartions
float BodyMetric::calculateBMI() const {
  return weight_ / (height_ * height_);
}

void BodyMetric::updateBMI() {
  bmi_ = calculateBMI();
}

string BodyMetric::determineBMIStatus() {
  if (bmi_ < 18.5) {
    bmiStatus_ = "Underweight";
  } else if (bmi_ < 25) {
    bmiStatus_ = "Normal weight";
  } else if (bmi_ < 30) {
    bmiStatus_ = "Overweight";
  } else if (bmi_ < 35) {
    bmiStatus_ = "Class I: Obesity";
  } else if (bmi_ < 40) {
    bmiStatus_ = "Class II: Obesity";
  }else {
    bmiStatus_ = "Class III: Obesity";
  }

  return bmiStatus_;
}

// Getters
int BodyMetric::getAge() const {
  return age_;
}

string BodyMetric::getGender() const {
  return gender_;
}

float BodyMetric::getWeight() const {
  return weight_;
}

float BodyMetric::getHeight() const {
  return height_;
}

string BodyMetric::getGoal() const {
  return goal_;
}

float BodyMetric::getBMI() const {
  return bmi_;
}

string BodyMetric::getBMIStatus() const {
  return bmiStatus_;
}

// Setters
void BodyMetric::setAge(int age) {
  age_ = age;
}

void BodyMetric::setGender(string gender) {
  gender_ = gender;
}

void BodyMetric::setWeight(float weight) {
  weight_ = weight;
  updateBMI(); // Updates BMI when weight is changed
}

void BodyMetric::setHeight(float height) {
  height_ = height;
  updateBMI(); // Updates BMI when weight is changed
}

void BodyMetric::setGoal(string goal) {
  goal_ = goal;
}
```

```python
%%writefile Person.h
#ifndef PERSON_H
#define PERSON_H

#include "BodyMetric.h"
#include <string>
using namespace std;

class Person {
private:
  string username_;
  string password_;
  BodyMetric bodyMetric_; // composition

public:

  // constructors
  Person();
  Person(string username, string password); // overloaded constructor

  // getters
  string getUsername() const;
  string getPassword() const;
  BodyMetric& getBodyMetric();
  const BodyMetric& getBodyMetric() const;

  // setters
  void setUsername(string username);
  void setPassword(string password);
  void setBodyMetric(const BodyMetric& bm);

};

#endif // PERSON_H
```

```python
%%writefile Person.cpp

#include "BodyMetric.h"
#include "Person.h"
using namespace std;

// Default Constructor
Person::Person() {
  username_ = "";
  password_ = "";
}

// Overloaded Constructor
Person::Person(string username, string password) {
  username_ = username;
  password_ = password;
}

// Getters
string Person::getUsername() const {
  return username_;
}

string Person::getPassword() const {
  return password_;
}

BodyMetric& Person::getBodyMetric() {
  return bodyMetric_;
}
const BodyMetric& Person::getBodyMetric() const {
  return bodyMetric_;
}

// Setters
void Person::setUsername(string username) {
  username_ = username;
}

void Person::setPassword(string password) {
  password_ = password;
}

void Person::setBodyMetric(const BodyMetric& bm) {
  bodyMetric_ = bm;
}
```

```python
%%writefile Database.h
#ifndef DATABASE_H
#define DATABASE_H

#include "Workout.h"
#include "BodyMetric.h"
#include "Person.h"
#include <string>
using namespace std;

const int MAX_USERS = 50; // maximum number of users allowed
const int MAX_WORKOUTS = 20; // maximum number of workouts allowed

class Database {
private:

  string credentialFile_;
  string userInfoFile_;
  int userCount_;
  Person person_[MAX_USERS]; // array of User objects
  int workoutCount_;
  Workout workout_[MAX_WORKOUTS]; // array of Workout objects

public:

  // Constructor
  Database(string credfile, string userfile);

  // loading / saving file to and from memory (credentials)
  void loadFromFile();
  void saveToFile();

  // loading / saving file to and from memory (userFile)
  void loadUserFile();
  void saveUserFile();

  // Load & save workout plans from file to memory
  void loadWorkoutPlans(const string& workoutFile);
  void saveWorkoutPlans(string workoutFile);

  // login and registration functions
  bool login(string username, string password) const;
  bool registerUser(string username, string password);

  // utilities
  int getUserIndex(string username) const; // returns the index of the user with the given username, or -1 if not found
  bool needsProfileUpdate(string username) const; // returns true if the user needs to update their profile, false otherwise
  void updateProfile(string username); // updates the user's profile with the given bodyMetric

  // health Info Submenu actions
  void displayProfile(int userIndex) const;
  void updateWeightHeight(int userIndex);
  void setWeightGoal(int userIndex);
  void suggestCalorieAndWorkout(int userIndex);
};

#endif // DATABASE_H
```

```python
%%writefile Database.cpp

#include "BodyMetric.h"
#include "Database.h"
#include <fstream>
#include <iostream>
#include <string>
#include <iomanip>

using namespace std;

// Constructor
Database::Database(string credfile, string userfile) {
  credentialFile_ = credfile;
  userInfoFile_ = userfile;
  userCount_ = 0;
}

// Load database.txt into memory
void Database::loadFromFile() {
  ifstream inFile(credentialFile_);

  if (!inFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << credentialFile_ << endl;
    return;
  }

  string username, password;
  while ((inFile >> username >> password) && userCount_ < MAX_USERS) { // Read username & password from file while userCount_ less than 50,
    person_[userCount_] = Person(username, password);                  // then initializing username & password in the Person constructor,
    userCount_++;                                                      // then initializing the person_ array with the objects.
  }
  inFile.close();
}

// load userInfo.txt into memory
void Database::loadUserFile() {
  ifstream inFile2(userInfoFile_);

  if (!inFile2.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << userInfoFile_ << "\n";
    return;
  }

  string uname, gender, goal;
  int age;
  float weight, height;

  while (inFile2 >> uname >> age >> gender >> weight >> height >> goal) { // Read all contents from file until end of file,
    for (int i = 0; i < userCount_; ++i) {                                // iterate through the file,
      if (person_[i].getUsername() == uname) {                            // match the username from database.txt & userInfo.txt,
        person_[i].getBodyMetric().setAge(age);                           // when match is found, health info from userInfo.txt is
        person_[i].getBodyMetric().setGender(gender);                     // correctly matched with the correct user and stored in
        person_[i].getBodyMetric().setWeight(weight);                     // person_ array. Now the array conatains, username, password
        person_[i].getBodyMetric().setHeight(height);                     // age, gender,weight,height and goal.
        person_[i].getBodyMetric().setGoal(goal);
        person_[i].getBodyMetric().updateBMI();
        person_[i].getBodyMetric().determineBMIStatus();
        break;  // stop searching once matched
      }
    }
  }
  inFile2.close();
}

// Load workout_plans.txt into memory
void Database::loadWorkoutPlans(const string& workoutFile) {
  ifstream inFile3(workoutFile);

  if (!inFile3.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << workoutFile << "\n";
    return;
  }

  workoutCount_ = 0;
  string activity, intensityStr, caloriesBurnStr;
  int intensityLvl, caloriesBurn;

  while (workoutCount_ < MAX_WORKOUTS           // While workoutCount_ is less than 20,
        && getline(inFile3, activity, ',')      // iterate through the file until delimiter is reached
        && getline(inFile3, intensityStr, ',')  // to extract data in csv format.
        && getline(inFile3, caloriesBurnStr))
  {
    intensityLvl  = stoi(intensityStr);     // Converts extracted string literals to integer(int) literals.
    caloriesBurn  = stoi(caloriesBurnStr);

    workout_[workoutCount_] = Workout(activity, intensityLvl, caloriesBurn); // Using the Workout constructor to initialize objects
    workoutCount_++;                                                         // to the workout_ array.
  }
  inFile3.close();
}

// Save database.txt from memory back to file
void Database::saveToFile(){
  ofstream outFile;
  outFile.open(credentialFile_);

  if (!outFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << credentialFile_ << endl;
    return;
  }

  for (int i = 0; i < userCount_; i++) {                                              // Iterate through the array and write
    outFile << person_[i].getUsername() << " " << person_[i].getPassword() << endl;   // back data to the file.
  }
  outFile.close();
}

// Save userInfo.txt from memory back to file
void Database::saveUserFile(){
  ofstream outFile;
  outFile.open(userInfoFile_);

  if (!outFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << userInfoFile_ << endl;
    return;
  }

  for (int i = 0; i < userCount_ ; i++) {                 // Iterate through the person_ array and
    outFile << person_[i].getUsername() << " "            // write back updated health info back to
        << person_[i].getBodyMetric().getAge() << " "     // userInfo.txt
        << person_[i].getBodyMetric().getGender() << " "
        << person_[i].getBodyMetric().getWeight() << " "
        << person_[i].getBodyMetric().getHeight() << " "
        << person_[i].getBodyMetric().getGoal() << endl;

  }
  outFile.close();
}

// Login Function
bool Database::login(string username, string password) const {
  for (int i = 0; i < userCount_; i++) {                                                 // Iterate through the person_ array
    if (person_[i].getUsername() == username && person_[i].getPassword() == password) {  // to find a match in database.txt else
      cout << "<Login successful!> Hello, " << person_[i].getUsername()  << endl;        // return false.
      return true;
    }
  }
  cout << "<Login failed!> Invalid username or password." << endl;
  return false;
}

// Registration Function
bool Database::registerUser(string username, string password) {                 // Warning when maximum number of users
  if (userCount_ >= MAX_USERS) {                                                // exceed 50
    cout << "<Registration failed!> Maximum number of users reached." << endl;
    return false;
  }

  for (int i = 0; i < userCount_; i++) {
    if (person_[i].getUsername() == username) {                                  // Warning when username already exists
      cout << "<Registration failed!> Username already exists." << endl;
      return false;
    }
  }

  person_[userCount_] = Person(username, password);                             // Register new user using Person Constructor
  userCount_++;
  cout << "<Registration successful!> Welcome, " << username << endl;
  return true;
}

// Utilities functions
// Retrive the index of the logged-in user
int Database::getUserIndex(string username) const {   // Retrive index of user from login
  for (int i = 0; i < userCount_; i++) {
    if (person_[i].getUsername() == username) {
      return i;
    }
  }
  return -1;
}

// Checks whether an account is new and prompts user to fill in health data
bool Database::needsProfileUpdate(string username) const {
  int index = getUserIndex(username);
  if (index == -1) {
    return false;
  }

  return (person_[index].getBodyMetric().getAge() == 0);
}

// prompts the new user to fill in health data
void Database::updateProfile(string username) {
    int index = getUserIndex(username);
    if (index == -1) {
        cout << "User not found.\n";
        return;
    }

    int age;
    string gender;
    float weight;
    float height;
    string goal;

    // Prompt & read
    cout << "Enter your age: ";
    cin  >> age;

    cout << "Enter your gender (M/F): ";
    cin  >> gender;

    cout << "Enter your weight (kg): ";
    cin  >> weight;

    cout << "Enter your height (m): ";
    cin  >> height;

    cout << "Enter your weight goal (lose/maintain/gain): ";
    cin  >> goal;

    // Build a new BodyMetric object and assign it to the person_ array
    BodyMetric bm(age, gender, weight, height, goal);
    person_[index].setBodyMetric(bm);

    cout << "Profile updated successfully.\n";
}

// health Info Submenu actions

// Dispaly profile information
void Database::displayProfile(int userIndex) const {
  cout << "Username: " << person_[userIndex].getUsername() << endl;
  cout << "Age: " << person_[userIndex].getBodyMetric().getAge() << endl;
  cout << "Gender: " << person_[userIndex].getBodyMetric().getGender() << endl;
  cout << "Weight: " << person_[userIndex].getBodyMetric().getWeight() << " kg" << endl;
  cout << "Height: " << person_[userIndex].getBodyMetric().getHeight() << " m" << endl;
  cout << "Weight Goal: " << person_[userIndex].getBodyMetric().getGoal() << endl;
  cout << "BMI: " << person_[userIndex].getBodyMetric().getBMI()
       << " (" << person_[userIndex].getBodyMetric().getBMIStatus() << ")" << endl << endl;
}

// Function to update weight & height of user
void Database::updateWeightHeight(int userIndex) {
  float weight, height;
  cout << "Enter your weight (kg): ";
  cin >> weight;
  cout << "Enter your height (m): ";
  cin >> height;

  person_[userIndex].getBodyMetric().setWeight(weight);
  person_[userIndex].getBodyMetric().setHeight(height);
  cout << "Weight and height updated successfully." << endl;
}

// Fucntion to aloow user update weight goal
void Database::setWeightGoal(int userIndex) {
  string goal;
  cout << "Enter your weight goal (Lose/Maintain/Gain): ";
  cin >> goal;
  person_[userIndex].getBodyMetric().setGoal(goal);
  cout << "Weight goal updated successfully." << endl;
}

// Function to suggest recommended calorie intake per day dan workout plans per day
void Database::suggestCalorieAndWorkout(int userIndex) {          // retrieve data
  float w = person_[userIndex].getBodyMetric().getWeight();
  float h = person_[userIndex].getBodyMetric().getHeight();
  int age = person_[userIndex].getBodyMetric().getAge();
  string goal = person_[userIndex].getBodyMetric().getGoal();
  string gender = person_[userIndex].getBodyMetric().getGender();
  float bmr;
  int calorieToBurn;

  if ( gender == "M" || gender == "m") {              // Calculate bmr & determine calorie cap based on gender of user
    bmr = (10 * w) + (625 * h) - (5 * age) + 5 ;
    calorieToBurn = 500;
  } else {
    bmr = (10 * w) + (625 * h) - (5 * age) - 161 ;
    calorieToBurn = 400;
  }

  int activityLevel;
  cout << "\nSelect your activity level:\n"                                 // Allow user to select multiplier to calculate TDEE
       << "1. Sedentary (little or no exercise)\n"                          // based on activity level of the user.
       << "2. Lightly active (light exercise/sports 1-3 days/week)\n"
       << "3. Moderately active (moderate exercise/sports 3-5 days/week)\n"
       << "4. Very active (hard exercise/sports 6-7 days/week)\n"
       << "5. Extra active (very hard exercise/sports and physical job)\n"
       << "Enter your choice: ";
  cin >> activityLevel;
  cout << endl;

  float multiplier;
  switch (activityLevel) { // TDEE Multiplier
    case 1:
      multiplier = 1.2;
      break;
    case 2:
      multiplier = 1.375;
      break;
    case 3:
      multiplier = 1.55;
      break;
    case 4:
      multiplier = 1.725;
      break;
    case 5:
      multiplier = 1.9;
      break;
    default:
      cout << "Invalid choice" << endl;
    break;
  }

  float tdee = bmr * multiplier; // Calculate TDEE based on multiplier selected by user

  float calorieIntake;                                       // Calculate calorie intake based on user's
  if (goal == "Lose" || goal == "lose") {                    // weight goal & TDEE
    calorieIntake = tdee - 500;
  } else if (goal == "Maintain" || goal == "maintain") {
    calorieIntake = tdee;
  } else {
    calorieIntake = tdee + 500;
  }

  cout << "TDEE (Total Daily Energy Expenditure): " << tdee << " calories" << endl;
  cout << "Suggested calorie intake based on weight goal: " << calorieIntake << " calories" << endl;

  // Suggest workout plan
  int workCount = 0;
  int suggestedWork[MAX_WORKOUTS]; //Array to store suggested workouts

  cout << "\nSuggested workouts:\n"; // Suggest workout based on weight goal with it's fixed inetsnity level, then save to the suggestWork array
  for (int i = 0; i < workoutCount_; i++) {
    if ((goal == "Lose" || goal == "lose") && workout_[i].getIntensityLevel() >= 3) {
      suggestedWork[workCount] = i;
      cout << workCount+1 << ". " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;

    } else if ((goal == "Maintain" || goal == "maintain") && workout_[i].getIntensityLevel() == 2) {
      suggestedWork[workCount] = i;
      cout << workCount+1 << ". " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;

    } else if ((goal == "Gain" || goal == "gain") && workout_[i].getIntensityLevel() <= 2) {
      suggestedWork[workCount] = i;
      cout << workCount+1 << ". " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;
    }
  }

  // Error message if no match is found
  if (workCount == 0) {
    cout << "No suitable workouts found for your goal." << endl;
    return;
  }

  // Allow user to pick one activity suggested and dispaly time taken minutes/day to reach burn cap
  int choice;
  cout << "\nEnter the number of the workout you want to do: ";
  cin >> choice;
  cout << endl;

  if (choice < 1 || choice > workCount) {
  cout << "Invalid selection.\n";
  return;
  }

  float timeNeeded = static_cast<float>(calorieToBurn) / workout_[suggestedWork[choice-1]].getCaloriesBurnPerMinute();
  cout << "You need to do " << workout_[suggestedWork[choice-1]].getActivity() << " for " << timeNeeded << " minutes/day to reach your goal." << endl;
}
```

```python
%%writefile main.cpp

#include "Database.h"
#include <iostream>

using namespace std;

int main(){

  const string credfile = "database.txt";
  const string userfile = "userInfo.txt";

  Database db(credfile, userfile);

  db.loadFromFile();
  db.loadUserFile();
  db.loadWorkoutPlans("workout_plans.txt");

  string username, password;
  int choice;

  do {
    cout << "****************************************" << endl;
    cout << "***** Welcome to My Health Journal *****" << endl;
    cout << "Menu: " << endl;
    cout << "[1] Login" << endl;
    cout << "[2] Register" << endl;
    cout << "[3] Exit" << endl;
    cout << "Enter your choice: ";
    cin >> choice;
    cout << endl;

    switch (choice) {
      case 1:
        cout << "****************************************" << endl;
        cout << "**************** Login *****************" << endl;
        cout << "Enter username: ";
        cin >> username;
        cout << "Enter password: ";
        cin >> password;

        // First Time login
        if(db.login(username, password)){ // Check if user is new
          if(db.needsProfileUpdate(username)){  // Redirect users to fill in their profile
            db.updateProfile(username); // Update user's profile
          }

          // Health Submenu
          int subChoice;
          int index = db.getUserIndex(username); // retrive index of user that logged-in
          do {
            cout << "****************************************" << endl;
            cout << "*********** Health Info Menu ***********" << endl;
            cout << "[1] Display Profile Info" << endl;
            cout << "[2] Update Weight & Height" << endl;
            cout << "[3] Set Weight Goal" << endl;
            cout << "[4] Suggest calorie Intake & workout plan" << endl;
            cout << "[5] Logout & return to Main Menu" << endl;
            cout << "Enter your choice: ";
            cin >> subChoice;
            cout << endl;

            switch (subChoice) {
              case 1:
                db.displayProfile(index);
                break;
              case 2:
                db.updateWeightHeight(index);
                break;
              case 3:
                db.setWeightGoal(index);
                break;
              case 4:
                db.suggestCalorieAndWorkout(index);
                break;
              case 5:
                cout << "Returning to Main Menu..." << endl;
                break;
              default:
                cout << "Invalid choice" << endl;
            }
          } while (subChoice != 5);
        }
        break;

      case 2:
        cout << "****************************************" << endl;
        cout << "*************** Register ***************" << endl;
        cout << "Enter username: ";
        cin >> username;
        cout << "Enter password: ";
        cin >> password;
        db.registerUser(username, password);
        break;

      case 3:
        cout << "Thanks for visiting!" << endl;
        break;

      default:
        cout << "Invalid choice" << endl;
    }

  } while (choice != 3);

  cout << "Part 2  " << endl;
  db.saveToFile();
  db.saveUserFile();

  return 0;
}
```
