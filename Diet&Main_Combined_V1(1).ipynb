```python
%%writefile BodyMetric.h
#ifndef BODYMETRIC_H // include guard
#define BODYMETRIC_H

#include <string> // for string class
using namespace std;

class BodyMetric {
private:
  int age_;
  string gender_;
  float weight_;
  float height_;
  string goal_;
  float bmi_;
  string bmiStatus_;

public:

  // constructors
  BodyMetric();
  BodyMetric(int age, string gender, float weight, float height, string goal);

  // BMI operations
  float calculateBMI() const;
  void updateBMI();
  string determineBMIStatus();

  // Getters
  int getAge() const;
  string getGender() const;
  float getWeight() const;
  float getHeight() const;
  string getGoal() const;
  float getBMI() const;
  string getBMIStatus() const;

  // Setters
  void setAge(int age);
  void setGender(string gender);
  void setWeight(float weight);
  void setHeight(float height);
  void setGoal(string goal);

};

#endif // BODYMETRIC_H
```

```python
%%writefile BodyMetric.cpp

#include <string>
#include "BodyMetric.h"
using namespace std;

// Default Constructor
BodyMetric::BodyMetric() {
  age_ = 0;
  gender_ = "";
  weight_ = 0.0;
  height_ = 0.0;
  goal_ = "";
}

// Overloaded Constructor
BodyMetric::BodyMetric(int age, string gender, float weight, float height, string goal) {
  age_ = age;
  gender_ = gender;
  weight_ = weight;
  height_ = height;
  goal_ = goal;
}

// BMI opeartions
float BodyMetric::calculateBMI() const {
  return weight_ / (height_ * height_);
}

void BodyMetric::updateBMI() {
  bmi_ = calculateBMI();
}

string BodyMetric::determineBMIStatus() {
  if (bmi_ < 18.5) {
    bmiStatus_ = "Underweight";
  } else if (bmi_ < 25) {
    bmiStatus_ = "Normal weight";
  } else if (bmi_ < 30) {
    bmiStatus_ = "Overweight";
  } else if (bmi_ < 35) {
    bmiStatus_ = "Class I: Obesity";
  } else if (bmi_ < 40) {
    bmiStatus_ = "Class II: Obesity";
  }else {
    bmiStatus_ = "Class III: Obesity";
  }

  return bmiStatus_;
}

// Getters
int BodyMetric::getAge() const {
  return age_;
}

string BodyMetric::getGender() const {
  return gender_;
}

float BodyMetric::getWeight() const {
  return weight_;
}

float BodyMetric::getHeight() const {
  return height_;
}

string BodyMetric::getGoal() const {
  return goal_;
}

float BodyMetric::getBMI() const {
  return bmi_;
}

string BodyMetric::getBMIStatus() const {
  return bmiStatus_;
}

// Setters
void BodyMetric::setAge(int age) {
  age_ = age;
}

void BodyMetric::setGender(string gender) {
  gender_ = gender;
}

void BodyMetric::setWeight(float weight) {
  weight_ = weight;
  updateBMI(); // Updates BMI when weight is changed
}

void BodyMetric::setHeight(float height) {
  height_ = height;
  updateBMI(); // Updates BMI when weight is changed
}

void BodyMetric::setGoal(string goal) {
  goal_ = goal;
}
```

```python
%%writefile DietPlans.h
#ifndef DIETPLANS_H
#define DIETPLANS_H
#include<iostream>
using namespace std;

const int MIN_F_INTAKE = 1200; // minimum calories intake for female
const int MIN_M_INTAKE = 1500; // minimum calories intake for male

class DietPlans {
private:
    // Original DietPlans data members
    string type_; // Breakfast, Lunch, Dinner, Snack
    string plan_; // Oatmeal with banana, Grilled chicken salad, Salmon and veggies, Yogurt with nuts
    int calories_;

    // Original DietFeatures data members
    int caloriesBurned_;
    DietPlans* dpArray_;

public:
    // Original DietPlans constructors and methods
    DietPlans();
    DietPlans(string, string, int);
    string getType() const;
    string getPlan() const;
    int getCalories() const;
    void setType(string);
    void setPlan(string);
    void setCalories(int);
    void setMeal(string, string, int);

    // Original DietFeatures constructors and methods
    DietPlans(int, int); // constructor with burned calories and numPlans
    ~DietPlans(); // destructor for dynamic memory
    int getCaloriesBurned() const;
    void setCaloriesBurned(int);
    string isRecorded(int);
    int selectPlan(const DietPlans*, int, string);
    void trackCalories(const DietPlans*, int[]);
    void calcCalDiff(int, int, char, int&);
    void defineSuggestion(const DietPlans[], int[], int, int);
    void suggestMultipleChange(const DietPlans&, const DietPlans&, int&, bool&);
    void suggestSingleChange(const DietPlans&, const DietPlans&, int&);
    DietPlans* getDpArray();
};

#endif
```

```python
%%writefile DietPlans.cpp

#include "DietPlans.h"
#include<iostream>
#include<string>
#include<iomanip>
using namespace std;

// ============= ORIGINAL DIETPLANS METHODS =============

DietPlans::DietPlans(){
    type_ = "";
    plan_ = "";
    calories_ = 0;
    caloriesBurned_ = 0;
    dpArray_ = nullptr;
}

DietPlans::DietPlans(string t, string p, int c){
    type_ = t;
    plan_ = p;
    calories_ = c;
    caloriesBurned_ = 0;
    dpArray_ = nullptr;
}

string DietPlans::getType() const{
    return type_;
}

string DietPlans::getPlan() const{
    return plan_;
}

int DietPlans::getCalories() const{
    return calories_;
}

void DietPlans::setType(string t){
    type_ = t;
}

void DietPlans::setPlan(string p){
    plan_ = p;
}

void DietPlans::setCalories(int c){
    calories_ = c;
}

void DietPlans::setMeal(string t, string p, int c){
    setType(t);
    setPlan(p);
    setCalories(c);
}

// ============= ORIGINAL DIETFEATURES METHODS =============

DietPlans::DietPlans(int burned, int numPlans) {
    type_ = "";
    plan_ = "";
    calories_ = 0;
    dpArray_ = new DietPlans[numPlans];
    caloriesBurned_ = burned;
}

DietPlans::~DietPlans() {
    delete[] dpArray_;
    dpArray_ = nullptr;
}

DietPlans* DietPlans::getDpArray() {
    return dpArray_;
}

int DietPlans::getCaloriesBurned() const {
    return caloriesBurned_;
}

void DietPlans::setCaloriesBurned(int calories) {
    caloriesBurned_ = calories;
}

//suggestMultipleChange()
// suggests multiple meal plan change to meet limit range
// &suggest: status for meal type change suggestion(if suggested to change = true)
void DietPlans::suggestMultipleChange(const DietPlans &lp, const DietPlans &op, int &diff, bool &suggest){
    //end the process if the looped plan calories is higher than the original plan
    if(op.getCalories() < lp.getCalories())
        return;

    diff -= op.getCalories() - lp.getCalories(); //calories difference of the current looped plan and original plan
    cout << setw(15) << left << "" << "Change " << lp.getType() << " to " << lp.getPlan() << endl;
    suggest = true;
}

//suggestSingleChange()
// only suggest changing 1 meal plan to meet limit range
// & for lp and op, used to refer original object, prevent new object creation
// lp: current looped meal plan
// op: original selected meal plan
void DietPlans::suggestSingleChange(const DietPlans &lp, const DietPlans &op, int &diff){
    int x = op.getCalories() - lp.getCalories(); //calories difference of the current looped plan and original plan
    if(x >= diff){
        cout << setw(15) << left << "" << "Change " << lp.getType() << " to " << lp.getPlan() << endl;
        diff = 0;
    }
}

//defineSuggestion()
// define and prints meal suggestions
void DietPlans::defineSuggestion(const DietPlans dp[], int plans[], int diff, int dpSize) {
    string mealTypes[4] = {"Breakfast", "Lunch", "Dinner", "Snack"};
    bool selected[4] = {plans[0] != -1, plans[1] != -1, plans[2] != -1, plans[3] != -1};

    if (diff < 0) {
        // Suggest adding a meal if under the minimum intake
        for (int i = 0; i < 4; ++i) {
            if (!selected[i]) {
                for (int j = 0; j < dpSize; ++j) {
                    if (dp[j].getType() == mealTypes[i] && dp[j].getCalories() <= -diff) {
                        cout << "Suggestion: Add a " << mealTypes[i] << " Meal: " << dp[j].getPlan() << endl;
                        return;
                    }
                }
            }
        }
    } else if (diff > 0) {
        // Suggest removing snack first if it exists
        if (selected[3]) {
            cout << setw(15) << left << "Suggestion: " << "Don't eat snacks." << endl;
            diff -= dp[plans[3]].getCalories();
        }

        // Try single meal change to fix over-intake
        for (int i = 0; i < 4; ++i) {
            if (!selected[i]) continue;
            for (int j = 0; j < dpSize; ++j) {
                if (dp[j].getType() == mealTypes[i] && j != plans[i]) {
                    suggestSingleChange(dp[j], dp[plans[i]], diff);
                    if (diff <= 0) return;
                }
            }
        }

        // If still over, suggest multiple changes
        bool suggested[4] = {false, false, false, false};
        for (int i = 0; i < dpSize; ++i) {
            string type = dp[i].getType();
            for (int m = 0; m < 4; ++m) {
                if (type == mealTypes[m] && selected[m] && i != plans[m] && !suggested[m]) {
                    suggestMultipleChange(dp[i], dp[plans[m]], diff, suggested[m]);
                    if (diff <= 0) return;
                }
            }
        }
    }
}

//calcCalDiff()
// calculates the difference of the total calories with the recomended range
void DietPlans::calcCalDiff(int tot, int limit, char gender, int &diff){
    int min = (gender == 'F')? MIN_F_INTAKE : MIN_M_INTAKE;

    if(tot < min){
        diff = tot - min; //gets a - value, to indicate it doesn't meet the min limit
        cout << "< Minimum calories intake not met >" << endl;
    }
    else if(tot > limit){
        diff = tot - limit; //gets + value, to indicate it exceed
        cout << "< Calories intake exceed recommended limit >" << endl;
    }
    else{
        diff = 0;
    }
}

//trackCalories()
// calculate and return the total calories intake of the selected meal plans
void DietPlans::trackCalories(const DietPlans dp[], int plan[]){
    int tot=0;
    for(int i = 0; i < 4; i++){
        if(plan[i] != -1){
            tot += dp[plan[i]].getCalories();
        }
    }
    setCaloriesBurned(tot);
}

//selectPlan()
// prints all the meal plans for a specific meal type
// then prompts user to select a meal plan
// returns the dpArray index of selected meal plan
int DietPlans::selectPlan(const DietPlans dp[], int plans, string meal){
    int opt = 1;    //plan option index
    int plan = 0;   //plan selection
    int j = 0;      //index for dpIndex
    int dpIndex[plans]; //array of the indexes of the particular meal types
    while(true){
        cout << meal << "Meal Plan: " << endl;

        for(int i = 0; i < plans; i++){
            if(dp[i].getType() == meal){
                cout << "[" << opt++ << "] " << dp[i].getPlan() << " (" << dp[i].getCalories() << " calories)" << endl;
                dpIndex[j++] = i;
            }
        }
        int maxOpt = opt - 1;
        cout << "Enter Plan: ";
        cin >> plan;
        if(plan > 0 && plan <= maxOpt){
            break;
        }
        else{
            cout << "Invalid input. Please enter a number between 1 and " << maxOpt << "." << endl;
            opt=1;
            j=0; // Reset j counter for retry
        }
    }
    return dpIndex[plan - 1];
}

//isRecorded()
// prints a (Recorded) behind the meal type in the menu to indicate that
// the meal plan for the meal type is already chosen
string DietPlans::isRecorded(int meal){
    if(meal == -1){
        return "";
    }
    else{
        return " (Recorded)";
    }
}
```

```python
%%writefile Database.h
#ifndef DATABASE_H
#define DATABASE_H

#include "Workout.h"
#include "BodyMetric.h"
#include "Person.h"
#include "DietPlans.h"
#include <string>
using namespace std;

const int MAX_USERS = 50; // maximum number of users allowed
const int MAX_WORKOUTS = 20; // maximum number of workouts allowed
const int NUM_PLANS = 20;      //number of plans

class Database {
private:

  string credentialFile_;
  string userInfoFile_;
  int userCount_;
  Person person_[MAX_USERS]; // array of User objects
  int workoutCount_;
  Workout workout_[MAX_WORKOUTS]; // array of Workout objects
  DietPlans* dp_; // array of Meal plans (pointer to array allocated in main)
  DietPlans df_;

public:

  // Constructor
  Database(string credfile, string userfile, DietPlans* dietPlans); // Added DietPlans* parameter

  // loading / saving file to and from memory (credentials)
  void loadFromFile();
  void saveToFile();

  // loading / saving file to and from memory (userFile)
  void loadUserFile();
  void saveUserFile();

  // Load & save workout plans from file to memory
  void loadWorkoutPlans(const string& workoutFile);
  void saveWorkoutPlans(string workoutFile);

  // login and registration functions
  bool login(string username, string password) const;
  bool registerUser(string username, string password);

  // utilities
  int getUserIndex(string username) const; // returns the index of the user with the given username, or -1 if not found
  bool needsProfileUpdate(string username) const; // returns true if the user needs to update their profile, false otherwise
  void updateProfile(string username); // updates the user's profile with the given bodyMetric

  // health Info Submenu actions
  void displayProfile(int userIndex) const;
  void updateWeightHeight(int userIndex);
  void setWeightGoal(int userIndex);
  void suggestCalorieAndWorkout(int userIndex);

  //Diet part of the program
  void selectMealPlan (int userIndex);
};

#endif // DATABASE_H
```

```python
%%writefile Database.cpp

#include "BodyMetric.h"
#include "Database.h"
#include <fstream>
#include <iostream>
#include <string>
#include <iomanip>

using namespace std;

// Constructor
Database::Database(string credfile, string userfile, DietPlans* dietPlans) { // Added DietPlans* parameter
  credentialFile_ = credfile;
  userInfoFile_ = userfile;
  userCount_ = 0;
  dp_ = dietPlans; // Assign the received array to the member variable
}

// Load database.txt into memory
void Database::loadFromFile() {
  ifstream inFile(credentialFile_);

  if (!inFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << credentialFile_ << endl;
    return;
  }

  string username, password;
  while ((inFile >> username >> password) && userCount_ < MAX_USERS) { // Read username & password from file while userCount_ less than 50,
    person_[userCount_] = Person(username, password);                  // then initializing username & password in the Person constructor,
    userCount_++;                                                      // then initializing the person_ array with the objects.
  }
  inFile.close();
}

// load userInfo.txt into memory
void Database::loadUserFile() {
  ifstream inFile2(userInfoFile_);

  if (!inFile2.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << userInfoFile_ << "\n";
    return;
  }

  string uname, gender, goal;
  int age;
  float weight, height;

  while (inFile2 >> uname >> age >> gender >> weight >> height >> goal) { // Read all contents from file until end of file,
    for (int i = 0; i < userCount_; ++i) {                                // iterate through the file,
      if (person_[i].getUsername() == uname) {                            // match the username from database.txt & userInfo.txt,
        person_[i].getBodyMetric().setAge(age);                           // when match is found, health info from userInfo.txt is
        person_[i].getBodyMetric().setGender(gender);                     // correctly matched with the correct user and stored in
        person_[i].getBodyMetric().setWeight(weight);                     // person_ array. Now the array conatains, username, password
        person_[i].getBodyMetric().setHeight(height);                     // age, gender,weight,height and goal.
        person_[i].getBodyMetric().setGoal(goal);
        person_[i].getBodyMetric().updateBMI();
        person_[i].getBodyMetric().determineBMIStatus();
        break;  // stop searching once matched
      }
    }
  }
  inFile2.close();
}

// Load workout_plans.txt into memory
void Database::loadWorkoutPlans(const string& workoutFile) {
  ifstream inFile3(workoutFile);

  if (!inFile3.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << workoutFile << "\n";
    return;
  }

  workoutCount_ = 0;
  string activity, intensityStr, caloriesBurnStr;
  int intensityLvl, caloriesBurn;

  while (workoutCount_ < MAX_WORKOUTS           // While workoutCount_ is less than 20,
        && getline(inFile3, activity, ',')      // iterate through the file until delimiter is reached
        && getline(inFile3, intensityStr, ',')  // to extract data in csv format.
        && getline(inFile3, caloriesBurnStr))
  {
    intensityLvl = stoi(intensityStr);     // Converts extracted string literals to integer(int) literals.
    caloriesBurn = stoi(caloriesBurnStr);

    workout_[workoutCount_] = Workout(activity, intensityLvl, caloriesBurn); // Using the Workout constructor to initialize objects
    workoutCount_++;                                                         // to the workout_ array.
  }
  inFile3.close();
}

// Save database.txt from memory back to file
void Database::saveToFile(){
  ofstream outFile;
  outFile.open(credentialFile_);

  if (!outFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << credentialFile_ << endl;
    return;
  }

  for (int i = 0; i < userCount_; i++) {                                              // Iterate through the array and write
    outFile << person_[i].getUsername() << " " << person_[i].getPassword() << endl;   // back data to the file.
  }
  outFile.close();
}

// Save userInfo.txt from memory back to file
void Database::saveUserFile(){
  ofstream outFile;
  outFile.open(userInfoFile_);

  if (!outFile.is_open()) { // Check for file opening errors
    cout << "Unable to open file: " << userInfoFile_ << endl;
    return;
  }

  for (int i = 0; i < userCount_ ; i++) {                 // Iterate through the person_ array and
    outFile << person_[i].getUsername() << " "            // write back updated health info back to
        << person_[i].getBodyMetric().getAge() << " "     // userInfo.txt
        << person_[i].getBodyMetric().getGender() << " "
        << person_[i].getBodyMetric().getWeight() << " "
        << person_[i].getBodyMetric().getHeight() << " "
        << person_[i].getBodyMetric().getGoal() << endl;

  }
  outFile.close();
}

// Login Function
bool Database::login(string username, string password) const {
  for (int i = 0; i < userCount_; i++) {                                                    // Iterate through the person_ array
    if (person_[i].getUsername() == username && person_[i].getPassword() == password) {     // to find a match in database.txt else
      cout << "<Login successful!> Hello, " << person_[i].getUsername()  << endl << endl;   // return false.
      return true;
    }
  }
  cout << "<Login failed!> Invalid username or password." << endl;
  return false;
}

// Registration Function
bool Database::registerUser(string username, string password) {                 // Warning when maximum number of users
  if (userCount_ >= MAX_USERS) {                                                // exceed 50
    cout << "<Registration failed!> Maximum number of users reached." << endl;
    return false;
  }

  for (int i = 0; i < userCount_; i++) {
    if (person_[i].getUsername() == username) {                                  // Warning when username already exists
      cout << "<Registration failed!> Username already exists." << endl;
      return false;
    }
  }

  person_[userCount_] = Person(username, password);                             // Register new user using Person Constructor
  userCount_++;
  cout << "<Registration successful!> Welcome, " << username << endl;
  return true;
}

// Utilities functions
// Retrive the index of the logged-in user
int Database::getUserIndex(string username) const {   // Retrive index of user from login
  for (int i = 0; i < userCount_; i++) {
    if (person_[i].getUsername() == username) {
      return i;
    }
  }
  return -1;
}

// Checks whether an account is new and prompts user to fill in health data
bool Database::needsProfileUpdate(string username) const {
  int index = getUserIndex(username);
  if (index == -1) {
    return false;
  }

  return (person_[index].getBodyMetric().getAge() == 0);
}

// prompts the new user to fill in health data
void Database::updateProfile(string username) {
  int index = getUserIndex(username);

  if (index == -1) {
    cout << "User not found.\n";
    return;
  }

  int age;
  string gender;
  float weight;
  float height;
  string goal;

  // Prompt & read
  cout << "+--------------------------------------+" << endl;
  cout << "|             Profile Update           |" << endl;
  cout << "+--------------------------------------+" << endl;
  cout << "Please enter your health information:\n";

  do {
    cout << "Enter your age (number): ";
    cin  >> age;
  } while ( age < 1 );

  do {
    cout << "Enter your gender (M/F): ";
    cin  >> gender;
  } while ( gender != "M" && gender != "F" && gender != "m" && gender != "f");

  do {
    cout << "Enter your weight (kg): ";
    cin  >> weight;
  } while ( weight < 0 );

  do {
    cout << "Enter your height (m): ";
    cin  >> height;
  } while ( height < 0);

  do {
    cout << "Enter your weight goal (lose/maintain/gain): ";
    cin  >> goal;
  } while ( goal != "Lose" && goal != "Maintain" && goal != "Gain" && goal != "lose" && goal != "maintain" && goal != "gain");

  // Build a new BodyMetric object and assign it to the person_ array
  BodyMetric bm(age, gender, weight, height, goal);
  person_[index].setBodyMetric(bm);

  cout << "Profile updated successfully.\n";
}

// health Info Submenu actions

// Dispaly profile information
void Database::displayProfile(int userIndex) const {

  cout << fixed << setprecision(2);
  cout << "+--------------------------------------+" << endl;
  cout << "|             Profile Info             |" << endl;
  cout << "+--------------------------------------+" << endl;
  cout << "Username: " << person_[userIndex].getUsername() << endl;
  cout << "Age: " << person_[userIndex].getBodyMetric().getAge() << endl;
  cout << "Gender: " << person_[userIndex].getBodyMetric().getGender() << endl;
  cout << "Weight: " << person_[userIndex].getBodyMetric().getWeight() << " kg" << endl;
  cout << "Height: " << person_[userIndex].getBodyMetric().getHeight() << " m" << endl;
  cout << "Weight Goal: " << person_[userIndex].getBodyMetric().getGoal() << endl;
  cout << "BMI: " << person_[userIndex].getBodyMetric().getBMI()
       << " (" << person_[userIndex].getBodyMetric().getBMIStatus() << ")" << endl << endl;

  char charChoice;
  do {
  cout << "Press [Y] to continue: ";
  cin >> charChoice;
  } while (charChoice != 'Y' && charChoice != 'y');
  cout << endl;
}

// Function to update weight & height of user
void Database::updateWeightHeight(int userIndex) {
  float weight, height;

  cout << "+--------------------------------------+" << endl;
  cout << "|         Update Weight & Height       |" << endl;
  cout << "+--------------------------------------+" << endl;

  do {
  cout << "Enter your weight (kg): ";
  cin >> weight;
  } while (weight < 0);

  do {
  cout << "Enter your height (m): ";
  cin >> height;
  } while (height < 0);

  person_[userIndex].getBodyMetric().setWeight(weight);
  person_[userIndex].getBodyMetric().setHeight(height);
  cout << "Weight and height updated successfully." << endl;

  char charChoice;
  do {
  cout << "Press [Y] to continue: ";
  cin >> charChoice;
  } while (charChoice != 'Y' && charChoice != 'y');
  cout << endl;
}

// Fucntion to aloow user update weight goal
void Database::setWeightGoal(int userIndex) {
  string goal;

  cout << "+--------------------------------------+" << endl;
  cout << "|            Set Weight Goal           |" << endl;
  cout << "+--------------------------------------+" << endl;
  do {
  cout << "Enter your weight goal (Lose/Maintain/Gain): ";
  cin >> goal;
  } while ( goal != "Lose" && goal != "Maintain" && goal != "Gain" && goal != "lose" && goal != "maintain" && goal != "gain");
  person_[userIndex].getBodyMetric().setGoal(goal);
  cout << "Weight goal updated successfully." << endl;

  char charChoice;
  do {
  cout << "Press [Y] to continue: ";
  cin >> charChoice;
  } while (charChoice != 'Y' && charChoice != 'y');
  cout << endl;

}

// Function to suggest recommended calorie intake per day dan workout plans per day
void Database::suggestCalorieAndWorkout(int userIndex) {          // retrieve data
  float w = person_[userIndex].getBodyMetric().getWeight();
  float h = person_[userIndex].getBodyMetric().getHeight();
  int age = person_[userIndex].getBodyMetric().getAge();
  string goal = person_[userIndex].getBodyMetric().getGoal();
  string gender = person_[userIndex].getBodyMetric().getGender();
  float bmr;
  int calorieToBurn;

  cout << fixed << setprecision(2);
  cout << "+--------------------------------------+" << endl;
  cout << "|     Suggested Calorie Intake         |" << endl;
  cout << "|         & Workout Plan               |" << endl;
  cout << "+--------------------------------------+" << endl;

  if ( gender == "M" || gender == "m") {              // Calculate bmr & determine calorie cap based on gender of user
    bmr = (10 * w) + (625 * h) - (5 * age) + 5 ;
    calorieToBurn = 500;
  } else {
    bmr = (10 * w) + (625 * h) - (5 * age) - 161 ;
    calorieToBurn = 400;
  }

  int activityLevel;
  cout << "+--------------------------------------------------------------------+" << endl;
  cout << "|                    Select your activity level:                     |" << endl;
  cout << "+--------------------------------------------------------------------+" << endl;   // Allow user to select multiplier to calculate TDEE
  cout << "|  [1] Sedentary (little or no exercise)                             |" << endl;   // based on activity level of the user.
  cout << "|  [2] Lightly active (light exercise/sports 1-3 days/week)          |" << endl;
  cout << "|  [3] Moderately active (moderate exercise/sports 3-5 days/week)    |" << endl;
  cout << "|  [4] Very active (hard exercise/sports 6-7 days/week)              |" << endl;
  cout << "|  [5] Extra active (very hard exercise/sports and physical job)     |" << endl;
  cout << "+--------------------------------------------------------------------+" << endl;
  do {
  cout << "Enter your choice: ";
  cin >> activityLevel;
  } while ( activityLevel < 1 || activityLevel > 5);
  cout << endl;

  float multiplier;
  switch (activityLevel) { // TDEE Multiplier
    case 1:
      multiplier = 1.2;
      break;
    case 2:
      multiplier = 1.375;
      break;
    case 3:
      multiplier = 1.55;
      break;
    case 4:
      multiplier = 1.725;
      break;
    case 5:
      multiplier = 1.9;
      break;
    default:
      cout << "Invalid choice" << endl;
    break;
  }

  float tdee = bmr * multiplier; // Calculate TDEE based on multiplier selected by user

  float calorieIntake;                                       // Calculate calorie intake based on user's
  if (goal == "Lose" || goal == "lose") {                    // weight goal & TDEE
    calorieIntake = tdee - 500;
  } else if (goal == "Maintain" || goal == "maintain") {
    calorieIntake = tdee;
  } else {
    calorieIntake = tdee + 500;
  }

  cout << "TDEE (Total Daily Energy Expenditure): " << tdee << " calories" << endl;
  cout << "Suggested calorie intake based on weight goal: " << calorieIntake << " calories" << endl << endl;

  // Suggest workout plan
  int workCount = 0;
  int suggestedWork[MAX_WORKOUTS]; //Array to store suggested workouts

  cout << "+--------------------------------------+" << endl;
  cout << "|          Suggested workouts          |" << endl; // Suggest workout based on weight goal with it's fixed inetsnity level, then save to the suggestWork array
  cout << "+--------------------------------------+" << endl;
  for (int i = 0; i < workoutCount_; i++) {
    if ((goal == "Lose" || goal == "lose") && workout_[i].getIntensityLevel() >= 3) {
      suggestedWork[workCount] = i;
      cout << "[" << workCount+1 << "] " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;

    } else if ((goal == "Maintain" || goal == "maintain") && workout_[i].getIntensityLevel() == 2) {
      suggestedWork[workCount] = i;
      cout <<"["<< workCount+1 << "] " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;

    } else if ((goal == "Gain" || goal == "gain") && workout_[i].getIntensityLevel() <= 2) {
      suggestedWork[workCount] = i;
      cout << "["<< workCount+1 << "] " << workout_[i].getActivity() << " (" << workout_[i].getIntensityLevel() << ") - " << workout_[i].getCaloriesBurnPerMinute() << " calories/minute\n";
      workCount++;
    }
  }

  // Error message if no match is found
  if (workCount == 0) {
    cout << "No suitable workouts found for your goal." << endl;
    return;
  }

  // Allow user to pick one activity suggested and dispaly time taken minutes/day to reach burn cap
  int choice;
  do {
  cout << "\nEnter the number of the workout you want to do: ";
  cin >> choice;
  } while ( choice < 1 || choice > workCount);
  cout << endl;

  float timeNeeded = static_cast<float>(calorieToBurn) / workout_[suggestedWork[choice-1]].getCaloriesBurnPerMinute();
  cout << "You need to do " << workout_[suggestedWork[choice-1]].getActivity() << " for " << static_cast<int>(timeNeeded) << " minutes/day to reach your goal." << endl << endl;

  char charChoice;
  do {
  cout << "Press [Y] to continue: ";
  cin >> charChoice;
  } while (charChoice != 'Y' && charChoice != 'y');
  cout << endl;
}

void Database :: selectMealPlan (int userIndex) {

  ifstream dietFile;
  dietFile.open("diet_plans.txt");

  if(!dietFile){ //Check if dietfile is open
    cout << "Error opening file" << endl;
    return;
  }

  string type = ""; //meal type
  string plan = ""; //meal plan(food plan)
  int calories = 0; //calories of meal
  int numPlans = 0;

  // Initialize df_ with the DietFeatures constructor parameters
  df_ = DietPlans(0, NUM_PLANS); // Using the combined class constructor

  while (getline(dietFile, type, ',') && getline(dietFile, plan, ',') && dietFile >> calories) {
    dietFile.ignore(); // to skip the newline character after reading calories
    dp_[numPlans++].setMeal(type, plan, calories);
  }

  dietFile.close(); //close file

  int mealtype = 0; //meal type option
  int dailyPlan[4] = {-1,-1,-1,-1}; //meal plan selection, stores the plan index according dpArray
  int totalCal = 0; //total calories intake
  //refer previous variables
  int calories_limit = 2000; //limit suggested by program
  int calories_diff = 0;
  string gender = person_[userIndex].getBodyMetric().getGender(); //gender of user
  char genderChar = gender.empty() ? 'M' : gender[0];
  int rec_cal = 0; //recommended calories intake

  do{
    cout << "**********************************" << endl;
    cout << "*********** Daily Diet ***********" << endl;
    cout << "Meal Type:" << endl;
    cout << "[1] Breakfast" << df_.isRecorded(dailyPlan[0]) << endl;
    cout << "[2] Lunch" << df_.isRecorded(dailyPlan[1]) << endl;
    cout << "[3] Dinner" << df_.isRecorded(dailyPlan[2]) << endl;
    cout << "[4] Snack" << df_.isRecorded(dailyPlan[3]) << endl;
    cout << "[5] Done" << endl;
    cout << "**********************************" << endl;
    cout << "Recommended Calories Intake: " << rec_cal << endl;
    df_.trackCalories(dp_, dailyPlan);
    cout << "Current Calories Intake: " << df_.getCaloriesBurned() << endl;
    totalCal= df_.getCaloriesBurned();
    df_.calcCalDiff(totalCal, calories_limit, genderChar, calories_diff);
    //print suggestions
    df_.defineSuggestion(dp_, dailyPlan, calories_diff, numPlans);
    cout << "Enter Meal Type: ";
    cin >> mealtype;

    cout << "**********************************" << endl;
    switch(mealtype){
      case 1:
        dailyPlan[0] = df_.selectPlan(dp_, numPlans, "Breakfast");
        break;
      case 2:
        dailyPlan[1] = df_.selectPlan(dp_, numPlans, "Lunch");
        break;
      case 3:
        dailyPlan[2] = df_.selectPlan(dp_, numPlans, "Dinner");
        break;
      case 4:
        dailyPlan[3] = df_.selectPlan(dp_, numPlans, "Snack");
        break;
      case 5:
        df_.trackCalories(dp_, dailyPlan);
        cout << "Total Calories Intake: " << df_.getCaloriesBurned() << " calories" << endl;
        break;
      default:
        cout << "Invalid option" << endl;
        break;
    }

  }while(mealtype != 5);
}
```
